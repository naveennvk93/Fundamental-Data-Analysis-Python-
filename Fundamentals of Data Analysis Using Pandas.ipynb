```python
# Method 2

ufo_cols = ['city', 'colors_reported', 'shape_reported', 'state', 'time', 'location']

ufo.columns = ufo_cols

ufo.head()
```

```python
# Renaming columns of the data set while reading the data file

ufo_cols = ['city', 'colors_reported', 'shape_reported', 'state', 'time']

ufo = pd.read_csv('http://bit.ly/uforeports', names = ufo_cols, header = 0)

ufo.head()
```

```python
# Method 4 - Replacing all the column names using string function

ufo = pd.read_csv('http://bit.ly/uforeports')

print(ufo.columns)

ufo.columns = ufo.columns.str.replace(' ', '_')

print(ufo.columns)
```

#### Groupby in Pandas

```python
drinks.head()
```

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
drinks.loc[drinks.continent == 'Africa', ['country','beer_servings']].sort_values('beer_servings').head()
```

```python
drinks.loc[drinks.continent == 'Africa', ['country','beer_servings']].beer_servings.agg(['sum','max','min'])
```

```python
drinks.groupby('continent').beer_servings.agg(['sum','max','min','mean'])
```

```python
drinks.groupby('continent').agg(['sum','max','min','mean'])
```

```python
%matplotlib inline
```

```python
drinks.groupby('continent').mean().plot(kind='bar')
```
